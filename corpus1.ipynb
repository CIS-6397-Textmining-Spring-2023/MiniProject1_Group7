### top 30 words from corpus-1 without removing stop words 

```python
import os
import glob
from collections import Counter
import matplotlib.pyplot as plt
import wordcloud
# Define the path to your corpus folder
corpus_folder = 'C:\\Users\\vasuk\\OneDrive\\Desktop\\sql\\Corpus1\\Corpus1'

# Function to read and tokenize a text file
def read_and_tokenize(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        words = text.split()
        return words

# Initialize a Counter to store word frequencies
word_counter = Counter()

# Iterate through all text files in the corpus folder
for file_path in glob.glob(os.path.join(corpus_folder, '*.txt')):
    words = read_and_tokenize(file_path)
    word_counter.update(words)

# Sort the word frequencies in descending order
word_freq = word_counter.most_common()

top_words = 30  # Number of top words to print
for i, (word, freq) in enumerate(word_freq):
    if i >= top_words:
        break
    print(f'{word}: {freq}')
# If you want to visualize the distribution, you can use a bar chart
top_words = 30  # Number of top words to visualize
top_words_data = word_freq[:top_words]

words, frequencies = zip(*top_words_data)

plt.figure(figsize=(12, 6))
plt.bar(words, frequencies)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title(f'Top {top_words} Words in the Corpus-1')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()




# code for wordcloud
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Create a dictionary from the words and frequencies
word_freq_dict = {word: freq for word, freq in zip(words, frequencies)}

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq_dict)

# Display the word cloud using Matplotlib
plt.figure(figsize=(7, 3))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# Remove stop words corpus -1 

```python
import os
import glob
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords

# Define the path to your corpus folder and the custom output folder
corpus_folder = 'C:\\Users\\vasuk\\OneDrive\\Desktop\\sql\\Corpus1\\Corpus1'
custom_output_folder = 'C:\\Users\\vasuk\\OneDrive\\Desktop\\sql\\stop_word_remove_corpus1'  # Create this folder if it doesn't exist

# Get a list of common English stop words
stop_words = set(stopwords.words('english'))

# Function to remove stop words from a given text
def remove_stop_words(text):
    words = nltk.word_tokenize(text)
    words = [word.lower() for word in words if word.lower() not in stop_words]
    return ' '.join(words)

# Process and save each text file in the corpus folder with custom output paths
for file_path in glob.glob(os.path.join(corpus_folder, '*.txt')):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    
    # Remove stop words from the text
    processed_text = remove_stop_words(text)
    
    # Define a custom output file path based on your preferences
    # Here, we use the original filename and place it in the custom output folder
    output_file = os.path.join(custom_output_folder, os.path.basename(file_path))
    
    # Save the processed text to the custom output file
    with open(output_file, 'w', encoding='utf-8') as output_file:
        output_file.write(processed_text)

print("Stop words removed from all files in the corpus folder and saved with custom output file paths.")
```

# pre processing text data

```python
import os
import glob
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')

# Define the path to your corpus folder and the output folder
corpus_folder = 'C:\\Users\\vasuk\\OneDrive\\Desktop\\sql\\Corpus1\\Corpus1'
output_folder = 'C:\\Users\\vasuk\\OneDrive\\Desktop\\sql\\pre_processed_corpus1'

# Ensure the output folder exists
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Initialize a Porter Stemmer
stemmer = PorterStemmer()

# Define a function to preprocess text
def preprocess_text(text):
    # Tokenize the text
    tokens = nltk.word_tokenize(text)
    
    # Lowercase the tokens
    tokens = [token.lower() for token in tokens]
    
    # Remove punctuation
    tokens = [token for token in tokens if token not in string.punctuation]
    
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    
    # Stem the tokens
    tokens = [stemmer.stem(token) for token in tokens]
    
    # Join the tokens back into a cleaned text
    cleaned_text = ' '.join(tokens)
    
    return cleaned_text

# Process and save each text file in the corpus folder
for file_path in glob.glob(os.path.join(corpus_folder, '*.txt')):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    
    # Preprocess the text
    preprocessed_text = preprocess_text(text)
    
    # Define the output file path
    output_file = os.path.join(output_folder, os.path.basename(file_path))
    
    # Save the preprocessed text to the output file
    with open(output_file, 'w', encoding='utf-8') as output_file:
        output_file.write(preprocessed_text)

print("Text data has been preprocessed and saved in the output folder.")
```
